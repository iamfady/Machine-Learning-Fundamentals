```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# loading The Data from csv file to pandas DataFram

iris_data = pd.read_csv('/content/drive/MyDrive/Data/Machine Learning Fundamentals/iris.csv')
```

```python
iris_data.head()
```

```python
iris_data['variety']
```

```python
iris_data['variety'].value_counts()
```

```python
label_encoder = LabelEncoder()
```

```python
iris_labels = label_encoder.fit_transform(iris_data.variety)
```

```python
iris_labels
```

```python
iris_data['target'] = iris_labels
```
